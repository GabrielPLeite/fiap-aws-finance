```python
import pandas as pd
```

# Ações

```python
df_acoes = pd.read_csv("fundamentus_acoes.csv")
```

```python
df_acoes.head()
```

```python
df_acoes.info()
```

```python
colunas_percent = [
    "DY","MargEBIT","MargLiq","ROIC","ROE","CrescRec5a"
]

colunas_numericas = [
    "Cotacao","PL","PVP","PSR","PAtivo","PCapGiro","PEBIT",
    "PAtivCircLiq","EVEBIT","EVEBITDA","LiqCorr",
    "Liq2meses","PatrimLiquido","DivBrutaPatrim"
]


for col in colunas_percent:
    df_acoes[col] = (
        df_acoes[col]
        .str.replace("%", "", regex=False)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
        .astype(float)
    ) / 100

for col in colunas_numericas:
    df_acoes[col] = (
        df_acoes[col]
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )
```

```python
df_acoes.info()
```

# FIIs

```python
df_fii = pd.read_csv("fundamentus_fii.csv")
```

```python
df_fii.head()
```

```python
df_fii.info()
```

```python
# colunas com porcentagem
colunas_percent = [
    "FFO_Yield",
    "Dividend_Yield",
    "Cap_Rate",
    "Vacancia_Media"
]

# colunas numéricas
colunas_numericas = [
    "Cotacao",
    "PVP",
    "Valor_de_Mercado",
    "Liquidez",
    "Preco_m2",
    "Aluguel_m2"
]

# tratar porcentagens
for col in colunas_percent:
    df_fii[col] = (
        df_fii[col]
        .str.replace("%", "", regex=False)
        .str.replace(",", ".", regex=False)
        .astype(float)
    ) / 100

# tratar números
for col in colunas_numericas:
    df_fii[col] = (
        df_fii[col]
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )
```

```python
df_fii.info()
```
